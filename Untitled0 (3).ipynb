```python
Q1. Explain Class and Object in OOP with an Example

Class: A class is a blueprint or template for creating objects. It defines properties (attributes) and behaviors (methods) that the objects created from it will have. Think of it as a custom data type that can contain data and functions to operate on that data.

Object: An object is an instance of a class. It represents a specific entity with the properties and behaviors defined by the class. Each object can have its own attribute values, which are defined by the class.


Example:

# Define a class
class Car:
    # Constructor to initialize attributes
    def _init_(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    # Method to display car information
    def display_info(self):
        print(f"{self.year} {self.make} {self.model}")

# Creating objects of Car class
car1 = Car("Toyota", "Camry", 2021)
car2 = Car("Honda", "Accord", 2020)

# Accessing the object's method
car1.display_info()  # Output: 2021 Toyota Camry
car2.display_info()  # Output: 2020 Honda Accord

In this example:

Car is the class.

car1 and car2 are objects created from the Car class.

display_info is a method defined in the class to print details about the car.



---

Q2. Name the Four Pillars of OOP

The four pillars of Object-Oriented Programming (OOP) are:

1. Encapsulation: Bundling of data and methods that operate on that data within a single unit or class, and restricting access to some of the object's components.


2. Abstraction: Hiding complex implementation details and only exposing essential features, allowing users to interact with objects at a high level.


3. Inheritance: Mechanism that allows one class to inherit attributes and methods from another class, promoting code reuse.


4. Polymorphism: Ability for functions, objects, or classes to operate in multiple forms, allowing the same interface to be used for different data types or classes.




---

Q3. Why the _init_() Function is Used? Give an Example

The _init_() function, also known as the constructor, is a special method in Python classes that is automatically called when an object is created. It is used to initialize the object’s attributes with user-defined values.

Example:

class Person:
    # Constructor to initialize attributes
    def _init_(self, name, age):
        self.name = name
        self.age = age

    # Method to display person information
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating an object of the Person class
person1 = Person("Alice", 30)

# Accessing the object's method
person1.display_info()  # Output: Name: Alice, Age: 30

In this example, _init_() initializes the name and age attributes when a Person object is created.


---

Q4. Why self is Used in OOP?

In Python, self is a reference to the current instance of the class. It is used as the first parameter in all instance methods to access the object's attributes and other methods within the class.

Without self, Python would not know which object’s attributes or methods you are referring to within the class.

Example:

class Dog:
    def _init_(self, name):
        self.name = name  # Using self to refer to the instance's attribute

    def bark(self):
        print(f"{self.name} says Woof!")

# Creating an object of Dog class
dog1 = Dog("Buddy")
dog1.bark()  # Output: Buddy says Woof!

In this example, self.name refers to the name attribute of the dog1 object.


---

Q5. What is Inheritance? Examples of Different Types of Inheritance

Inheritance is an OOP concept that allows a class (child class) to inherit attributes and methods from another class (parent class). It promotes code reuse by enabling the child class to inherit and extend the functionality of the parent class.

Types of Inheritance:

1. Single Inheritance: Involves one child class inheriting from one parent class.

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        print("Dog barks")

dog = Dog()
dog.speak()  # Output: Animal speaks
dog.bark()   # Output: Dog barks


2. Multiple Inheritance: Involves one child class inheriting from multiple parent classes.

class Father:
    def skill(self):
        print("Father's skill")

class Mother:
    def talent(self):
        print("Mother's talent")

class Child(Father, Mother):  # Child inherits from both Father and Mother
    pass

child = Child()
child.skill()   # Output: Father's skill
child.talent()  # Output: Mother's talent


3. Multilevel Inheritance: Involves a class inheriting from another class, which in turn inherits from another class.

class Vehicle:
    def type(self):
        print("This is a vehicle")

class Car(Vehicle):  # Car inherits from Vehicle
    def wheels(self):
        print("Car has 4 wheels")

class SportsCar(Car):  # SportsCar inherits from Car
    def brand(self):
        print("Sports car brand")

sportscar = SportsCar()
sportscar.type()   # Output: This is a vehicle
sportscar.wheels() # Output: Car has 4 wheels
sportscar.brand()  # Output: Sports car brand


4. Hierarchical Inheritance: Involves multiple child classes inheriting from the same parent class.

class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        print("Dog barks")

class Cat(Animal):  # Cat inherits from Animal
    def meow(self):
        print("Cat meows")

dog = Dog()
cat = Cat()
dog.sound()  # Output: Animal sound
cat.sound()  # Output: Animal sound
dog.bark()   # Output: Dog barks
cat.meow()   # Output: Cat meows


5. Hybrid Inheritance: A combination of multiple inheritance types, for example, combining multiple and multilevel inheritance.

In Hybrid Inheritance, multiple inheritance types are combined, which can create a complex class hierarchy. It typically involves a mix of multiple and multilevel inheritance structures, allowing a child class to inherit from multiple parents or ancestors in various ways.

For example:

class Animal:
    def sound(self):
        print("Animal makes a sound")

class Mammal(Animal):
    def has_fur(self):
        print("Mammal has fur")

class Bird(Animal):
    def has_feathers(self):
        print("Bird has feathers")

class Bat(Mammal, Bird):  # Bat inherits from both Mammal and Bird
    def fly(self):
        print("Bat can fly")

# Creating an object of Bat
bat = Bat()
bat.sound()        # Output: Animal makes a sound
bat.has_fur()      # Output: Mammal has fur
bat.has_feathers() # Output: Bird has feathers
bat.fly()          # Output: Bat can fly

In this example, Bat inherits from both Mammal and Bird, while Mammal and Bird both inherit from Animal. This creates a hybrid structure.

Key Points of I…
```
